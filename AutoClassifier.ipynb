```python
import numpy as np
import json

from keras.applications.vgg16 import VGG16
from keras.preprocessing.image import img_to_array, load_img
from keras.applications.imagenet_utils import preprocess_input
```

```python
vgg16=VGG16(weights='imagenet')
```

```python
vgg16.save('vgg16.h5')
```

```python
def prepare_image(img_path):
    img = load_img(img_path, target_size=(224, 224))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

```python
y=prepare_image('C:\Sem_6_Miniproject\ToyotaFortuner.jpg')
```

```python
preds=vgg16.predict(y)
```

```python
CLASS_INDEX=None
```

```python
import json
CLASS_INDEX_PATH='C:\Sem_6_Miniproject\imagenet_class_index.json'
```

```python
def get_predictions(preds,top=5):
  global CLASS_INDEX

  #Loading the json file
  CLASS_INDEX=json.load(open(CLASS_INDEX_PATH))

  #GETTING THE RESULTS
  results=[]
  for pred in preds:
    top_indices=pred.argsort()[-top:][::-1]
    result=[tuple(CLASS_INDEX[str(i)])+ (pred[i],) for i in top_indices]
    result.sort(key=lambda x:x[2], reverse=True)
    results.append(result)

  return results

```

```python
print(get_predictions(preds,top=5))
```

```python
from collections import Counter,defaultdict
import os
```

```python
import pickle as pk
```

```python
def get_car_categories():
    d=defaultdict(float)
    img_list=os.listdir('C:\\Sem_6_Miniproject\\archive\\cars_train\\cars_train')
    for i,img_path in enumerate(img_list):
        img=prepare_image('C:\\Sem_6_Miniproject\\archive\\cars_train\\cars_train\\'+img_path)
        out=vgg16.predict(img)
        top=get_predictions(out,top=5)
        for j in top[0]:
            d[j[0:2]]+=j[2]
            if i%100==0:
                print(i,'/',len(img_list),'complete')
    return Counter(d)
```

```python
cat_counter=get_car_categories()
```

```python
with open('cat_counter.pk','wb')as f:
    pk.dump(cat_counter,f,-1)
```

```python
#FIRST CHECK OF WHETHER IT's A CAR OR NOT
def car_categories_gate(image_path,cat_list):
    img=prepare_image(image_path)
    out=vgg16.predict(img)
    top=get_predictions(out,top=5)
    print("Validating that this is a picture of car")
    for j in top[0]:
        if j[0:2] in cat_list:
            print(j[0:2])
            return "Validation complete-Proceed to Damage Evaluation"
    return "Are you sure whether the uploaded image is that of a Car?"
```
